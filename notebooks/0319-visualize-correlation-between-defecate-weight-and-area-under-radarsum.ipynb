## Corrleation between defecation weight change and radar sum

- 0319
    - Manually exclude outliers by looking at the scatterplot. Fit linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_to_excel = '../data/processed/correlation-radar-sum-defecate-weight.csv'

res_df = pd.read_csv(path_to_excel, index_col=0)
```

```python
x = res_df['area_under_radarsum']
y = res_df['weight_change']

# manually exclude some "outlier" (done hastily)
# fit a linear regression
excluded = x > 30
x_ = x[~excluded]
y_ = y[~excluded]
m, b = np.polyfit(x_, y_, 1)
xs = np.linspace(0, max(x), num=1000)
ys = m * xs + b

# scatterplot and a fitted straight line
plt.figure(figsize = (6, 6))
plt.plot(x, y, '.')
plt.plot(xs, ys)
plt.xlabel('Area under radar_sum', size = 12)
plt.ylabel('Defecation Weight Change', size = 12)
plt.savefig('../reports/figures/corrleation-between-defecate-and-radar.png')
pass
```
